```python
column_names = ['Player Name','Nationality']
for i in range(1970,2020):
    column_names.append('year '+str(i))
```

```python
df2 = pd.DataFrame(columns=column_names)
```

```python
df2
```

```python
players[0] = "de villiers"
```

```python
for count,player in enumerate(players[100:200]):
    print(count,player)
    wd.get("http://stats.espncricinfo.com/ci/engine/stats/analysis.html")
    
    
    search_player = wd.find_element_by_class_name('guruInput')
    search_player.send_keys(player)
    
    search_button = wd.find_element_by_name('submit_analysis')
    search_button.click()
    
    players_link = wd.find_element_by_partial_link_text('Players and Officials')
    players_link.click()
    
    print('page loaded')
    
    
    try:
        pg_source1 = wd.page_source
        soup1 = BeautifulSoup(pg_source1, 'lxml')

        td1 = soup1.find(text='One-Day Internationals player').parent.parent
        tdp = td1.previous_siblings

        nationality = next(islice(tdp, 1,2)).text.split('\n')[1]
        odi_links = wd.find_elements_by_link_text('One-Day Internationals player')
        odi_links[0].click()

        submit_query = wd.find_element_by_name('submit_results')
        submit_query.click()

        print('details page loaded')

        page_source = wd.page_source

        soup = BeautifulSoup(page_source, 'lxml')
        years = soup.body.find_all(text=re.compile('year'))[:-1]
        runs = {'Player Name':player,'Nationality':nationality}

        for year in years:
            year_td = year.parent.parent
            next_tds = year_td.next_siblings
            try:
                runs[year] = (int(next(islice(next_tds, 5,6)).text))
            except:
                runs[year] = 0


       
    except:
        runs = {'Player Name':player}
        print('Problem fetching runs')
    print(runs)
    df2 = df2.append(runs, ignore_index = True)
    print(player+ ' record inserted successfully')
```
